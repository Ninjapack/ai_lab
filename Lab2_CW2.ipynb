```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
xs = np.random.uniform(low=-10, high=10, size=(1000, 1))
zs = np.random.uniform(low=-10, high=10, size=(1000, 1))
true_w1 = 2.0
true_w2 = -3.0
true_b = 5.0
targets = true_w1 * xs + true_w2 * zs + true_b
inputs = np.concatenate((xs, zs), axis=1)
np.savez('data.npz', inputs=inputs, targets=targets)
print("Dane zapisane.")
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.SGD(),
    loss=tf.keras.losses.MeanSquaredError()
)
```

```python
model.fit(inputs, targets, epochs=100, verbose=1)
```

```python
predictions = model.predict(inputs)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(predictions, label='Predykcje', alpha=0.7)
plt.plot(targets, label='Cele (targets)', alpha=0.7)
plt.legend()
plt.title("Porównanie predykcji i targetów")
plt.show()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(5, activation='sigmoid'),
    tf.keras.layers.Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse'
)

model.fit(inputs, targets, epochs=100, verbose=1)
```
